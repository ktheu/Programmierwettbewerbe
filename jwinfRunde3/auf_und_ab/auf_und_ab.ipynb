#### Schrittweise Lösung

Wir merken uns die Leiterübergänge in einem dictionary. Zunächst nur die von unten nach oben:

```python
leiter = {6:27, 14:19, 21:53, 31:42, 33:38, 46:62, 51:59, 57:96, 65:85, 68:80, 70:76, 92:98}
```

Wir fügen die Übergänge von oben nach unten hinzu

```python
for x in leiter.copy():        # da wir in der schleife leiter verändern, müssen wir eine copy durchlaufen
    leiter[leiter[x]] = x
print(leiter)
```

```python
wurf = 6           # Die Würfelzahl, die wir ausprobieren

anzahl = 0         # Anzahl Würfe
besucht = set()    # die schon besuchten Felder

pos = 0            # die aktuelle Position
while True:
    
    pos += wurf      # die neue Position
    anzahl +=1       # einmal mehr geworfen
    if pos == 100:   # Erfolg
        print(f'Mit {wurf} kommt man mit {anzahl} Würfen ans Ziel.')
        break
    elif pos > 100:  # zurückgehen
        zuviel = pos-100
        pos = 100-zuviel
        print(f'Beim {anzahl}-ten Wurf zurück auf {pos}.')

    if pos in leiter: # Leiter hoch oder runter
        pos = leiter[pos]

    if pos in besucht: # sind wir in einer Schleife gelandet?
        print(f'{pos} wird doppelt besucht. Mit {wurf} kommt man mit nicht ans Ziel.')
        break   
    besucht.add(pos)
```

#### Das ganze Programm

```python
wurf = 4             # Die Würfelzahl, die wir ausprobieren
print(f'Würfelzahl: {wurf}')

leiter = {6:27, 14:19, 21:53, 31:42, 33:38, 46:62, 51:59, 57:96, 65:85, 68:80, 70:76, 92:98}
for x in leiter.copy():        # da wir in der schleife leiter verändern, müssen wir eine copy durchlaufen
    leiter[leiter[x]] = x
    
anzahl = 0         # Anzahl Würfe
besucht = set()    # die schon besuchten Felder

pos = 0            # die aktuelle Position
while pos < 100:
    
    pos += wurf      # die neue Position
    anzahl +=1       # einmal mehr geworfen
    if pos == 100:   # Erfolg
        print(f'Mit {wurf} kommt man mit {anzahl} Würfen ans Ziel.')
        break
    elif pos > 100:  # zurückgehen
        zuviel = pos-100
        pos = 100-zuviel
        print(f'Beim {anzahl}-ten Wurf zurück auf {pos}.')

    if pos in leiter: # Leiter hoch oder runter
        pos = leiter[pos]

    if pos in besucht: # sind wir in einer Schleife gelandet?
        print(f'{pos} wird doppelt besucht. Mit {wurf} kommt man mit nicht ans Ziel.')
        break   
    besucht.add(pos)
```
