#### Lösung 

```python
# Einlesen der Daten
nr = '1'
print(f'Beispiel {nr}:')
f = open(f'beispieldaten/baywatch{nr}.txt')
georg = f.readline().split()
karte = f.readline().split()
f.close()
print(f'Georg: {georg}')
print(f'Karte: {karte}')
```

```python
def schieb(a, k):
    '''
    a: Liste,
    k: schiebeZahl, 0 <= k < len(a)
    returns: Liste, um k verschoben, d.h. die vorderen k Elemente wurden 
        nacheinander nach hinten versetzt.
    '''
    return [a[ (i+k) % len(a)] for i in range(len(a))]
```

```python
schieb([1,2,3,4,5],2)
```

```python
def passt(a,b):
    '''
    a: Liste mit Fragezeichen an manchen Stellen, b: Liste
    returns True, falls a gleich b an allen nicht-Fragezeichen Stellen
    ''' 
    for x,y  in zip(a,b):
        if x != '?' and x != y: return False
    return True
```

```python
passt([1,'?','?',4],[1,2,3,4])
```

```python
zaehl = 0
print('Karte:')
print(*karte)
for k in range(len(karte)):
    if passt(karte, schieb(georg,k)):
        zaehl+=1
        print('Verschiebung',k)
        print(*schieb(georg,k))

if zaehl == 0:
    print(f'Es gibt {zaehl} Lösungen.')
else:
    print(f'Es gibt {zaehl} Lösungen.')
```

#### Das ganze Programm

```python
def schieb(a, k):
    '''
    a: Liste,
    k: schiebeZahl, 0 <= k < len(a)
    returns: Liste, um k verschoben, d.h. die vorderen k Elemente wurden 
        nacheinander nach hinten versetzt.
    '''
    return [a[ (i+k) % len(a)] for i in range(len(a))]

def passt(a,b):
    '''
    a: Liste mit Fragezeichen an manchen Stellen, b: Liste
    returns True, falls a gleich b an allen nicht-Fragezeichen Stellen
    ''' 
    for x,y  in zip(a,b):
        if x != '?' and x != y: return False
    return True

nr = '2'
print(f'Beispiel {nr}:')
f = open(f'beispieldaten/baywatch{nr}.txt')
georg = f.readline().split()
karte = f.readline().split()
f.close()

zaehl = 0
print('Karte:')
print(*karte)
for k in range(len(karte)):
    if passt(karte, schieb(georg,k)):
        zaehl+=1
        print('Verschiebung',k)
        print(*schieb(georg,k))

if zaehl == 0:
    print(f'Es gibt keine Lösungen.')
elif zaehl == 1:
    print(f'Es gibt genau eine Lösung.')
else:
    print(f'Es gibt {zaehl} Lösungen.')
```
