#### Beispieldaten

Jede Datei beschreibt einen Zeitplan der Klassen und Unterrichtsstunden und ist so aufgebaut:

- Erste Zeile: Anzahl n der Unterrichtsstunden.
- Folgende n Zeilen: Jede Zeile enthält diese durch Leerzeichen getrennten Werte einer Unterrichtsstunde:
  Klassenbezeichner,  Wochentag, Startzeit, Endzeit, Klassengröße

```python
# Anschauen der Beispieldaten 00-07
nr = '00'
print(f'Beispiel {nr}:')
f = open(f'beispieldaten/ball{nr}.txt')
print(f.read().strip())
f.close()
```

#### Lösung

```python
# Einlesen der Daten
nr = '00'
print(f'Beispiel {nr}:')
f = open(f'beispieldaten/ball{nr}.txt')
n = int(f.readline())

data = []
for i in range(n):
    klasse, tag, von, bis, anzahl = f.readline().split()
    von = int(von)
    bis = int(bis)
    anzahl = int(anzahl)
    data.append((klasse, tag, von, bis, anzahl))
f.close()

print(f'{n=}')
print(f'{data=}')
```

Für jeden Tag erstellen wir eine Liste, in der wir uns für jede Stunde merken wir können, wieviele Bälle gebraucht werden.
Zu Beginn werden keine Bälle gebraucht.

```python
montag = [0]*24
dienstag = [0]*24   
mittwoch = [0]*24
donnerstag = [0]*24
freitag = [0]*24
```

```python
montag
```

Wir gehen die tupel in der Liste data durch und erhöhen an den entsprechenden Tagen und Stunden die Anzahl der benötigten Bälle.

```python
for klasse, tag, von, bis, anzahl in data:
    for t in range(von, bis):
        if tag == 'Montag':
            montag[t] += anzahl
        elif tag == 'Dienstag':
            dienstag[t] += anzahl
        elif tag == 'Mittwoch':
            mittwoch[t] += anzahl
        elif tag == 'Donnerstag':
            donnerstag[t] += anzahl
        elif tag == 'Freitag':
            freitag[t] += anzahl
```

```python
montag
```

Wir suchen die höchste Zahl in den Listen und merken uns, an welchem Tag und zu welcher Stunde die vorgekommen ist.

```python
best_tag = None
best_t = None
best_anzahl = 0
for t in range(24):
    if montag[t] > best_anzahl:
        best_anzahl = montag[t]
        best_tag = 'Montag'
        best_t = t
    if dienstag[t] > best_anzahl:
        best_anzahl = dienstag[t]
        best_tag = 'Dienstag'
        best_t = t
    if mittwoch[t] > best_anzahl:
        best_anzahl = mittwoch[t]
        best_tag = 'Mittwoch'
        best_t = t
    if donnerstag[t] > best_anzahl:
        best_anzahl = donnerstag[t]
        best_tag = 'Donnerstag'
        best_t = t
    if freitag[t] > best_anzahl:
        best_anzahl = freitag[t]
        best_tag = 'Freitag'
        best_t = t

print(f"Am {best_tag} um {best_t}:00 Uhr werden {best_anzahl} Bälle benötigt.")
```
