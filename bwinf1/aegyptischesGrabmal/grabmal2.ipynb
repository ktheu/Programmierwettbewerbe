```python
from heapq import heappop, heappush
def greedy(s):
    frontier = [s]
    prev = {s:None}
    while frontier:
        state = heappop(frontier)  
        if goaltest(state):
            return prev,state
        for v in nextstates(state):
            if v not in prev:
                heappush(frontier,v)
                prev[v] = state
    return None, None


def reconstructPath(prev,goalstate):
    state = goalstate
    path = []
    while state is not None:
        path.append(state)
        state = prev[state]
    path.reverse()
    return path
```

```python
'''
state = (t,n), n = Nummer des Abschnitts, 0 = noch außerhalb des Abschnitts, t = 1... Zeit
'''

def zustand(t,p):
    '''
    t = zeit, p = Periode
    returns: Zustand 'Z' = 0 zu oder 'A' = 1 auf und die Anzahl der zukünftigen Intervalle mit demselben Zustand
    '''
    return  0 if ((t - 1) // p) % 2 == 0 else 1

def nextstates(state):
    t,n = state
    tmp = []
    if n == 0:
        tmp.append((t+1,0))    # man kann weiter draußen (=Abschnitt 0) stehenbleiben
        if zustand(t,a[1]) == 1:
            tmp.append((t,1))
    if n == 1:
        tmp.append((t,0))    # man kann jederzeit vom Abschnitt 1 wieder zurück
        if zustand(t+1,a[1]) == 1:
            tmp.append((t+1,1))
        if zustand(t,a[2]) == 1:
            tmp.append((t,2))
    for k in range(2,len(a)-1):
        if n == k:
            if zustand(t+1,a[k]) == 1:
                tmp.append((t+1,k))
            if zustand(t,a[k+1]) == 1:
                tmp.append((t,k+1))
            if zustand(t,a[k-1]) == 1:
                tmp.append((t,k-1))
        
    return tmp
    
def goaltest(state):
    t,n = state
    return n == len(a)-1
```

```python
def t_change(path,t0):
    '''
    betrachtet path[t0-1:] 
    returns t >= t0, nach dem die Raumnummer im path[t0-1:] sich ändert
    '''
    i = 0
    while i < len(path) and path[i][0] < t0:
        i+=1

    if i >= len(path):
        return len(path)
    n = path[i][1]
    while i < len(path) and path[i][1] == n:
        i+=1
    return path[i][0]


def n_change(path,t):
    '''
    returns von wo nach wo zum Zeitpunkt t gewechselt wird
    '''
    i = 0
    while i < len(path) and path[i][0] < t:
        i+=1
    tvon, von = path[i]
    while i < len(path) and path[i][0] == tvon:
        i+=1
    
    nach = path[i][1] if i < len(path) else len(a)-1
    return von, nach
    

```

```python
%%time
nr = 5
print(f'Beispiel {nr}:')
f = open('grabmal'+str(nr)+'.txt')
n = int(f.readline())
a = [0]
for i in range(n):
    a.append(int(f.readline()))
  
startstate = (1,0)
prev, goalstate = greedy(startstate)
path = reconstructPath(prev,goalstate) 
ankunft = path[-1][0]


t = 0
while t < ankunft:
    t = t_change(path,t+1)
    von, nach = n_change(path,t)
    if von == 0:
        von = 'aussen'
    if nach == len(a)-1:
        nach = 'Grabmal'
    print(f'Minute {t} von {von} nach {nach}')
print()    
    
```
