```python
def zustand(p,t):
    '''
    p = periode, t = zeit
    returns: Zustand 'Z' = 0 zu oder 'A' = 1 auf und die Anzahl der zukünftigen Intervalle mit demselben Zustand
    '''
    return  0 if ((t - 1) // p) % 2 == 0 else 1
 
```

```python
from collections import deque
def bfs(startstate):
    frontier =  deque([startstate])
    prev = {startstate:None}
    while frontier:
        state = frontier.popleft()
        if goaltest(state):
            return prev, state
        for v in nextstates(state):
            if v not in prev:
                frontier.append(v)
                prev[v] = state
    return None, None


def reconstructPath(prev,goalstate):
    state = goalstate
    path = []
    while state is not None:
        path.append(state)
        state = prev[state]
    path.reverse()
    return path
```

```python
'''
state = (n,t), n = Nummer des Abschnitts, 0 = noch außerhalb des Abschnitts, t = 1... Zeit
'''

def nextstates(state):
    n,t = state
    tmp = []
    if n == 0:
        tmp.append((0,t+1))    # man kann weiter draußen (=Abschnitt 0) stehenbleiben
        if zustand(a[1],t) == 1:
            tmp.append((1,t))
    if n == 1:
        tmp.append((0,t))    # man kann jederzeit vom Abschnitt 1 wieder zurück
        if zustand(a[1],t+1) == 1:
            tmp.append((1,t+1))
        if zustand(a[2],t) == 1:
            tmp.append((2,t))
    for k in range(2,len(a)-1):
        if n == k:
            if zustand(a[k],t+1) == 1:
                tmp.append((k,t+1))
            if zustand(a[k+1],t) == 1:
                tmp.append((k+1,t))
            if zustand(a[k-1],t) == 1:
                tmp.append((k-1,t))
        
    return tmp
    
def goaltest(state):
    n,t = state
    return n == len(a)-1
```
