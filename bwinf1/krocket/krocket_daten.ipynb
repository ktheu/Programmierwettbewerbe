```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def show(nr):
    print('Beispiel:',nr)
    f = open('krocket'+str(nr)+'.txt')
    anz, radius = [int(x) for x in f.readline().split()]

    for i in range(anz):
         x1, y1, x2, y2 = [int(x) for x in f.readline().split()]
         plt.plot([x1,x2],[y1,y2])

    plt.show()
```
