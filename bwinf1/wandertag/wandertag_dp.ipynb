```python
def dp(i):
    # maximale Anzahl von Intervalle, die ich mit 3 Startpunkten aus a[i:] abdecken kann.
    # Original-Problem: dp(0)
    if i == n:
        return 0

    # punkt i wird nicht genommen
    val1 = dp(i+1)

    # punkt i wird genommen
    val2 = anzahl_intervalle_ueberdeckt_von(i)+ dp(i+1)  # aber dp(i+1) nur noch mit 2 Startpunkten
```

```python
def dp(i, k):
    # maximale Anzahl von Intervalle, die ich mit k Startpunkten aus a[i:] abdecken kann.
    # Original-Problem: dp(0,3)
    
    if i == n or k == 0:
        return 0

    # punkt i wird nicht genommen
    val1 = dp(i+1, k)

    # punkt i wird genommen
    val2 = anzahl_intervalle_ueberdeckt_von(i)+ dp(i+1, k-1)  # aber hier könnten Intervalle doppelt gezählt werden
```

```python
def dp(i, k, covered):
    # maximale Anzahl von Intervalle, die ich mit k Startpunkten aus a[i:] abdecken kann
    # wenn ich die Intervalle in coverered schon abgedeckt habe
    # Original-Problem: dp(0,3,leer)
    
    if i == n or k == 0:
        return len(covered)

    # punkt i wird nicht genommen
    val1 = dp(i+1, k, covered)

    # punkt i wird genommen
    iv = intervalle_ueberdeckt von(i)
    val2 = dp(i+1, k-1, covered vereinigt mit iv)  
```
