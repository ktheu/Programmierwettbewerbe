Jede Datei enthält eine Liste mit Musikstück und Figuren und ist so aufgebaut:

    - 1. Zeile: Eine ganze, positive und durch 8 teilbare Zahl m, die Anzahl der Takte des Musikstücks.
    - 2. Zeile: Die Anzahl n der aufgelisteten Figuren.
    - Folgende n Zeilen: Die verfügbaren Figuren bestehend aus diesen durch Leerzeichen getrennten Werten:
        1. Der Name der Figur.
        2. Die Anzahl der benötigten Takte.    
        3. Die Endposition der mit A bis P bezeichneten Tanzenden nach Ausführung der Figur,
          wenn sie vor der Figur in der Reihenfolge AB … OP standen.

Hier eine Beispieleingabe:
    
    32
    2
    Cast_Four 2 DABCHEFGLIJKPMNO
    Cast_Up 4 BCDEFGHIJKLMNOPA

In diesem Beispiel besteht das Musikstück aus 32 Takten und es stehen zwei Figuren zur Verfügung. Die Figur Cast_Four benötigt zwei Takte und nach der Ausführung stehen die Tanzenden in der Reihenfolge DABCHEFGLIJKPMNO (nach Ausgangsreihenfolge ABCDEFGHIJKLMNOP). Die Figur Cast_Up benötigt vier Takte und bringt die Tanzenden in die Reihenfolge BCDEFGHIJKLMNOPA (nach Ausgangsreihenfolge ABCDEFGHIJKLMNOP).

Bei der Beispieleingabe choreoA.txt können die Figuren mehr als 12 Takte haben. Sie ist daher keine Pflichteingabe, sondern als Herausforderung zu sehen.

#### Lösungsidee

Wir können immer wieder entscheiden, welche Figur wir als nächstes wählen wollen. Eine Lösung ist eine Abfolge von Entscheidungen, die uns zur Ausgangsstellung zurückführt. Wir suchen alle Lösungen, um uns dann nach den verschiedenen Kriterien die beste herauszusuchen.
Dazu nutzen wir Backtracking. Beim Backtracking bauen wir schrittweise eine Liste mit Entscheidungen auf und wenn wir merken, dass eine Entscheidung nicht zum Ziel führen kann, nehmen wir die zurück und prüfen die nächste Möglichkeit (den nächsten Kandidaten).

```python
def back(a):
    if isSolution(a):
        solutions.append(a.copy())
        return                       
    for cand in candidates(a):
        a.append(cand)
        back(a)
        a.pop()
                    
def isSolution(a):
    '''
    returns: True, wenn die Teillösung a eine Lösung darstellt
    '''
    return a[-1] == goal

def candidates(a):
    '''
    returns: Liste von Entscheidungen, die die Teillösung a um eine Stufe erweitern können
    '''
    s, k = a[-1]
    tmp = []
    for p, k1 in perms:
        if k+k1 <= m:
            tmp.append((apply(s,p),k+k1))
    return tmp

solutions = []
back([start])
print(len(solutions))
sol = solutions[0]
print(sol)
```

#### Das ganze Programm

```python
nr = '04'
print(f'Beispiel {nr}:')
f = open(f'beispieldaten/choreo{nr}.txt')
m = int(f.readline())   # Anzahl Takte des Musikstücks
n = int(f.readline())   # Anzahl gelisteter Figuren
figuren = [] 
for i in range(n):
    name, takte, s = f.readline().split()
    takte = int(takte)
    figuren.append((name,takte,list(s)))
 
f.close()
print(f'Beispiel choreo{nr}:')
print(f'Anzahl Takte des Musikstücks = {m}')
print(f'Anzahl Figuren = {n}')

names = dict()
perms = []
for figur in figuren:
    name, takte, s = figur
    p = tuple([ord(c)-65 for c in s])
    perms.append((p,takte))
    names[(p,takte)] = name


def apply(s, p):
    return [s[p[i]] for i in range(16)]

def back(a):
    if isSolution(a):
        solutions.append(a.copy())
        return                       
    for cand in candidates(a):
        a.append(cand)
        back(a)
        a.pop()
                    
def isSolution(a):
    '''
    returns: True, wenn die Teillösung a eine Lösung darstellt
    '''
    return a[-1] == goal

def candidates(a):
    '''
    returns: Liste von Entscheidungen, die die Teillösung a um eine Stufe erweitern können
    '''
    s, k = a[-1]
    tmp = []
    for p, k1 in perms:
        if k+k1 <= m:
            tmp.append((apply(s,p),k+k1))
    return tmp

def getMove(s1, s2):
    p1, k1 = s1
    p2, k2 = s2
    p = getPermutation(p1, p2)
    k = k2-k1
    return names[(p, k)]

def getMoves(path):
    moves = []
    for i in range(len(path)-1):
        moves.append(getMove(path[i],path[i+1]))
    return moves

def getPermutation(a, b):
    return tuple([a.index(c) for c in b])

def analyzePath(path):
    laenge = len(path)-1
    figuren = set()
    d = 0
    for i in range(len(path)-1):
        figuren.add(getMove(path[i], path[i+1]))
        d += distanz(path[i], path[i+1])
    return (laenge, len(figuren), d)

def distanz(s1, s2):
    p1, _ = s1
    p2, _ = s2
    return sum([abs(i-p2.index(p1[i])) for i in range(16)])

#Aufruf
from time import perf_counter
starttime = perf_counter()
solutions = []
start = (list('ABCDEFGHIJKLMNOP'), 0)
goal = (list('ABCDEFGHIJKLMNOP'), m)
back([start])
if len(solutions) == 0:
    print('Keine Lösung gefunden')
    endtime = perf_counter()
    print(f'Laufzeit: {endtime-starttime:.3f} Sekunden')

else:
    print(f'Anzahl gefundender Lösungen: {len(solutions)}')
    print()
    
    best_unterschiedliche_figuren = 0
    best_viele_figuren = 0
    best_wenige_figuren = float('inf')
    best_max_distanz = 0
    best_min_distanz = float('inf')
    best_unterschiedliche_figuren_path = None
    best_viele_figuren_path = None
    best_wenige_figuren_path = None
    best_max_distanz_path = None
    best_min_distanz_path = None
    for path in solutions:
        laenge, anz_figuren, d = analyzePath(path)
        if anz_figuren > best_unterschiedliche_figuren:
            best_unterschiedliche_figuren = anz_figuren
            best_unterschiedliche_figuren_path = path
        if laenge > best_viele_figuren:
            best_viele_figuren = laenge
            best_viele_figuren_path = path
        if laenge < best_wenige_figuren:
            best_wenige_figuren = laenge
            best_wenige_figuren_path = path
        if d > best_max_distanz:
            best_max_distanz = d
            best_max_distanz_path = path
        if d < best_min_distanz:
            best_min_distanz = d
            best_min_distanz_path = path
    
    print(f'Die meisten unterschiedlichen Figuren: {best_unterschiedliche_figuren }')
    print(' '.join(getMoves(best_unterschiedliche_figuren_path)))
    print(f'\nDie meisten Figuren: {best_viele_figuren}')
    print(' '.join(getMoves(best_viele_figuren_path)))
    print(f'\nDie wenigsten Figuren: {best_wenige_figuren}')
    print(' '.join(getMoves(best_wenige_figuren_path)))
    print(f'\nDie größte Distanz: {best_max_distanz}')
    print(' '.join(getMoves(best_max_distanz_path)))
    print(f'\nDie kleinste Distanz: {best_min_distanz}')
    print(' '.join(getMoves(best_min_distanz_path)))
    endtime = perf_counter()
    print(f'\nLaufzeit: {endtime-starttime:.3f} Sekunden')
 
```
