### Funktioniert nur in Sage-Umgebung

```python
def getGraph(nr):
    f = open('./beispieldaten/muellabfuhr'+str(nr)+'.txt')
    anzV, anzE = [int(x) for x in f.readline().split()]
    anzV, anzE
    G = {i:dict() for i in range(anzV)}
    for i in range(anzE):
        von, bis, km = [int(x) for x in f.readline().split()]
        G[von][bis]=km
        G[bis][von]=km
    return G
```

```python
G = getGraph(0)
Graph(G).show(edge_labels=True)
```
