```python
import matplotlib.pyplot as plt
import numpy as np
import random
%matplotlib inline
```

#### Punkte darstellen

```python
X = [1, 2, 3, 4]
Y = [1, 4, 9, 16]
 
C = ['red','red','blue','blue']
plt.figure(figsize=(4,3))
plt.scatter(X, Y, c=C, s=10)   # c = farbe, s = dicke

plt.show()
```

#### Darstellung einer Funktion

```python
x = np.arange(-5., 5., 0.2)

plt.figure(figsize=(4,3))
plt.plot(x, x**3, 'red')   # y = x^3
plt.xlabel("x")
plt.grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
plt.show()
```

#### Rechtecke in einem Koordinatensystem darstellen

```python
from matplotlib.patches import Rectangle

fig, ax = plt.subplots()
ax.set(xlim=(-50, 50), ylim=(-30, 30))
ax = plt.gca()
ax.axis('equal')  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
 
r1 = Rectangle((0,0),15,20)        # x,y,breite,höhe
r1.fill = False
r1.set_lw(2.0)
r1.set_color('blue')

r2 = Rectangle((-20,10),30,5)      # x,y,breite,höhe
r2.fill = False
r2.set_lw(2.0)
r2.set_color('red')
 
ax.add_patch(r1)
ax.add_patch(r2)
  
plt.grid(True)
```

#### Punkte diesseits und jenseits einer Geraden einfärben

```python
anzahl = 50
w1, w2, b = 1, -3, 0.5

X1 = [5*random.random() for _ in range(anzahl)] 
X2 = [5*random.random() for _ in range(anzahl)] 
C = [w1*X1[i]+ w2*X2[i] + b > 0 for i in range(anzahl)]

plt.figure(figsize=(4,3))
plt.scatter(X1, X2, c=C, s=10)

plt.axis([0, 5, 0, 5])
plt.show()
```

#### Punkte in einem Koordinatensystem darstellen

```python
points = [(-4,1),(-3,3),(-5,2),(-2,2),(-6,-2),(-4,-2),(1,1),(1,-1),(7,5)]

# Mit Grid und Koordinatenachsen
xcoords = [p[0] for p in points]
ycoords = [p[1] for p in points]
plt.plot(xcoords, ycoords, 'ob')     # o = Punkt, b = blue

ax = plt.gca()
ax.axis('equal')                  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.grid(True)

plt.show()
```

#### Ein Polygon in einem Koordinatensystem darstellen

```python
points = [(0,0), (200,0), (200,250), (150,100)]
# Mit Grid und Koordinatenachsen
xcoords = [p[0] for p in points]
ycoords = [p[1] for p in points]
plt.plot(xcoords, ycoords, 'ob')     # o = Punkt, b = blue
plt.plot(xcoords+[points[0][0]], ycoords+[points[0][1]])

ax = plt.gca()
ax.axis('equal')                  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.grid(True)

plt.show()
```

#### Einen Kreis einzeichnen

```python
points = [(0,0), (200,0), (200,250), (150,100)]
from matplotlib.patches import Circle
# Mit Grid und Koordinatenachsen
xcoords = [p[0] for p in points]
ycoords = [p[1] for p in points]
plt.plot(xcoords, ycoords, 'ob')     # o = Punkt, b = blue
plt.plot(xcoords+[points[0][0]], ycoords+[points[0][1]])

ax = plt.gca()
ax.axis('equal')                  
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
ax.add_patch(Circle((150, 100), 85, alpha=.2, color='red'))
plt.grid(True)

plt.show()
```
