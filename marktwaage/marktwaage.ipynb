#### Beispieldaten (0-5)

Jede Datei beschreibt einen Satz von Gewichtsstücken und enthält

  * in der ersten Zeile die Anzahl n der vorkommenden verschiedenen Gewichtsmaße
  * in den weiteren m Zeilen jeweils zwei Zahlen:
    * das Gewicht w der jeweiligen Gewichtsstücke in Gramm und
    * die Anzahl n_w der Gewichtsstücke mit diesem Gewicht.

Wir addieren die Gewichte, wenn wir sie auf die linke Seite legen, wir subtrahieren die Gewichte, wenn wir sie auf die rechte Seite legen. Ein 500g Stück auf die eine Seite und 10g und 50g auf die andere Seite entspricht also:

```
500 - 10 - 50 
```
Wir nennen einen solchen Ausdruck eine Gewichtssumme.
Wir vereinbaren, dass wir den zu wiegenden Gegenstand (target) immer auf die rechte Seite legen wollen. Unser Ziel ist es dann, eine Gewichtssumme zu konstruieren, die möglichst nah bei dem target-Gewicht liegt.



Wir ordnen die Gewichte der Reihe nach in eine Liste

````python
f = open('beispieldaten/gewichtsstuecke0.txt')
anz = int(f.readline())

gewichte = []
for i in range(anz):
    g, n = [int(x) for x in f.readline().split()]
    for i in range(n):
        gewichte.append(g)
        
gewichte
````

````python
def sel(i, target):
    '''
    returns: Liste mit bester Verteilung der Gewichte bis einschließlich Index i
    Wir stellen uns vor, das target wird links auf die Waagschale gelegt. Wenn wir 
    rechts ein Gewicht platzieren wollen, nehmen wir die positive Zahl in die Liste auf,
    wenn wir links ein Gewicht plaziere wollen, nehmen wir die negative Zahl in die Liste auf.
    
    Um Seiteneffekte zu vermeiden, geben wir immer Kopien von Listen zurück (es sei denn, die Liste wurde in 
    der Funktion explizit neu erstellt).
    '''
    if (i,target) in memo:
        return memo[(i, target)].copy()
    
    if i == -1: return []

    teil = gewichte[:i+1]   # diese gewichte dürfen wir platzieren
    
    # Wenn das target schwerer ist als alle Gewichte in teil, plazieren wir alle rechts.
    if target >= sum(teil):
        return teil
    
    # Wenn auf der rechten Seite etwas liegt, was schwerer als alle verfügbaren Gewichte ist, dann
    # plazieren wir alles nach links
    if -target >= sum(teil):
        return [-x for x in teil]
    
    
    # Für das letzte Gewicht gibt es drei Möglichkeiten:
    # linke Seite, rechte Seite oder nicht benutzen
    gewicht = gewichte[i]
    l1 = sel(i-1, target+gewicht)  # auf die linke Seite
    l1.append(-gewicht)
    
    l2 = sel(i-1, target)         # nicht benutzen
    
    l3 = sel(i-1, target-gewicht)  # auf die rechte Seite
    l3.append(gewicht)
    
    # der Fall r3 zeigt, dass unser rekursiver Aufruf auch negative targets erhalten kann:
    # wenn gewichte[i-1] sehr groß ist. Wir berücksichten dies oben.
    
    auswahl = [l1, l2, l3]

    t1 = sum(l1)
    t2 = sum(l2)
    t3 = sum(l3)
    
    vergleich = [(abs(t1-target), len(l1)), (abs(t2-target),
                                             len(l2)), (abs(t3-target), len(l3))]
    mindiff = min(vergleich)
    index = vergleich.index(mindiff)

    memo[(i, target)] = auswahl[index].copy()
    return auswahl[index].copy()
````

````python
%%time
f = open('beispieldaten/gewichtsstuecke0.txt')
anz = int(f.readline())

gewichte = []
for i in range(anz):
    g, n = [int(x) for x in f.readline().split()]
    for i in range(n):
        gewichte.append(g)
        
memo =dict()
zaehl = 0
for target in range(10,10001,10):
    a = sel(len(gewichte)-1,target)
    if target == sum(a):
        zaehl+=1
    
    print(target,sum(a),sel(len(gewichte)-1,target))
print('Anzahl exakte Lösungen: ', zaehl)
````
